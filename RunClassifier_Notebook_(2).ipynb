```python

# To load the dictionary back from the JSON file:
with open(working_folder + 'DATA/description_mapping.json', 'r') as file:
    description_mapping = json.load(file)
unlabeled_data = pd.read_pickle(working_folder + 'DATA/unlabeled_data_pickle')
model_data = pd.read_pickle(working_folder + 'DATA/model_data_pickle')
text_varname = 'comment'
labels_varname = 'numeric_labels'
```

```python
from src.ClassifierBERT import ClassifierBERT
```

```python
#!pip install tensorflow-gpu

import tensorflow as tf

if tf.test.gpu_device_name():
    print("GPU is available")
else:
    print("No GPU found")

```

```python

# Instantiate the BERT Classifier
MyClassifier = ClassifierBERT(model_data, text_varname, labels_varname, model='bert-base-uncased')
```

```python
# Train only non-BERT layers first:
MyClassifier.freeze_BERT_layers(unfreeze=False)
MyClassifier.train_model(epochs=15, batch_size=32, train_size=0.7, test_size=0.15, val_size=0.15,
                         initial_learning_rate=3e-5, data_augmentation=False, N=int(len(model_data)*0.1))
# Save trained model:
MyClassifier.save_model(path_folder + 'TextClassification_Python/DATA/Trained_Classifier_SHORT_v1')
```

```python

# Train the model with all layers:
MyClassifier.freeze_BERT_layers(unfreeze=True)
MyClassifier.train_model(epochs=10, batch_size=32, train_size=0.7, test_size=0.15, val_size=0.15,
                         initial_learning_rate=3e-5, data_augmentation=False, N=int(len(model_data)*0.1))
# Save trained model:
MyClassifier.save_model(path_folder + 'TextClassification_Python/DATA/Trained_Classifier_SHORT')
```

```python

# Load model:
MyClassifier_loaded = ClassifierBERT.load_model(path_folder + 'TextClassification_Python/DATA/Trained_Classifier_SHORT')
```

```python
# Make predictions on unlabeled data
unlabeled_data_unique = unlabeled_data[['comment','group_id']].drop_duplicates(subset='group_id', keep='first', inplace=False)
unlabeled_text_varname = 'comment'
unlabeled_data_unique = unlabeled_data_unique.sample(500)

df_human_readable_labels = MyClassifier_loaded.predict_unlabeled_data(unlabeled_data_unique, unlabeled_text_varname, description_mapping)

```
